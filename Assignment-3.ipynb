5. Given an integer number n, return the difference between the product of its digits and the sum of its digits.
   * Example 1:

Input: n = 234
Output: 15 
Explanation: 
Product of digits = 2 * 3 * 4 = 24 
Sum of digits = 2 + 3 + 4 = 9 
Result = 24 - 9 = 15

```python
import math

num1 = 234

# Convert number to string
num_str = str(num1)

# Convert each digit to int
digits = [int(d) for d in num_str]

# Product
product = math.prod(digits)
print("Product of digits:", product)

# Sum 
sum_of_digits = 0
for d in digits:
    sum_of_digits += d
print("Sum of digits:", sum_of_digits)

#Difference
result = product - sum_of_digits
print("Difference (Product - Sum):", result)
```
